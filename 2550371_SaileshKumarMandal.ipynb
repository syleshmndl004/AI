```python
#task 6 (Add HDI Category Column)
```

```python
# Function to classify HDI based on UNDP thresholds
def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi < 0.700:
        return "Medium"
    elif 0.700 <= hdi < 0.800:
        return "High"
    else:
        return "Very High"

# Apply classification
hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(classify_hdi)

# Verify the new column
hdi_2022_df[['country', 'hdi', 'HDI Category']].head()
```

```python
#task 3 (outlier detection(iqr method))
```

```python
# Function to detect outliers using 1.5 × IQR rule
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

# Detect outliers in HDI and GNI per Capita
south_asia_df['HDI_Outlier'] = detect_outliers(south_asia_df['hdi'])
south_asia_df['GNI_Outlier'] = detect_outliers(south_asia_df['gross_inc_percap'])

# Combine outliers
south_asia_df['Outlier'] = south_asia_df['HDI_Outlier'] | south_asia_df['GNI_Outlier']
```

```python
# Scatter plot highlighting outliers
plt.figure(figsize=(8,5))
plt.scatter(
    south_asia_df['gross_inc_percap'],
    south_asia_df['hdi'],
    c=south_asia_df['Outlier']
)
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Outlier Detection in South Asia (HDI vs GNI)")
plt.show()
```
