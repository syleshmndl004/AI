```python


def find_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return lower, upper

# Calculate bounds for Score
score_lower, score_upper = find_outliers(south_asia_df["score"])

# Calculate bounds for GDP
gdp_lower, gdp_upper = find_outliers(south_asia_df["Log GDP per capita"])

# Identify outliers
south_asia_df["Score_Outlier"] = (south_asia_df["score"] < score_lower) | (south_asia_df["score"] > score_upper)
south_asia_df["GDP_Outlier"] = (south_asia_df["Log GDP per capita"] < gdp_lower) | (south_asia_df["Log GDP per capita"] > gdp_upper)

# Display outlier countries
outliers = south_asia_df[(south_asia_df["Score_Outlier"]) | (south_asia_df["GDP_Outlier"])]
print("Outlier Countries in South Asia:")
outliers[["Country name","score","Log GDP per capita","Score_Outlier","GDP_Outlier"]]
```

```python
#Problem 3
#4. Happiness Disparity:

def range_and_cv(series):
    rng = series.max() - series.min()
    cv = series.std() / series.mean()
    return rng, cv

rng_sa, cv_sa = range_and_cv(south_asia_df["score"])
rng_me, cv_me = range_and_cv(middle_east_df["score"])

print("South Asia  Range:", rng_sa, "CV:", cv_sa)
print("Middle East  Range:", rng_me, "CV:", cv_me)

if cv_sa > cv_me:
    print("South Asia has higher variability.")
else:
    print("Middle East has higher variability.")
```


Which region has greater variability in happiness?

Based on the calculations, South Asia has a Coefficient of Variation (CV) of
approximately 0.3021482883337427, while the Middle East has a CV of approximately
0.28938880812119405. Therefore, South Asia shows slightly greater relative variability in
happiness scores compared to the Middle East, meaning the happiness scores in
South Asia are more spread out relative to their mean.

```python

def iqr_bounds(s):
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    return q1 - 1.5*iqr, q3 + 1.5*iqr

s_low, s_high = iqr_bounds(south_asia_df["score"])
g_low, g_high = iqr_bounds(south_asia_df["Log GDP per capita"])

south_asia_df["Score_Outlier"] = (south_asia_df["score"] < s_low) | (south_asia_df["score"] > s_high)
south_asia_df["GDP_Outlier"]   = (south_asia_df["Log GDP per capita"] < g_low) | (south_asia_df["Log GDP per capita"] > g_high)

outliers = south_asia_df[south_asia_df["Score_Outlier"] | south_asia_df["GDP_Outlier"]]
```

```python
#Problem 3
#6. Outlier Detection:
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR

    return df[(df[col] < lower) | (df[col] > upper)]
```

```python
#outliers in score

sa_score_out = detect_outliers(south_asia_df, "score")
me_score_out = detect_outliers(middle_east_df, "score")

print("South Asia Score Outliers:")
display(sa_score_out)

print("Middle East Score Outliers:")
display(me_score_out)
```

```python
#outliers in gdp

sa_gdp_out = detect_outliers(south_asia_df, "Log GDP per capita")
me_gdp_out = detect_outliers(middle_east_df, "Log GDP per capita")

print("South Asia GDP Outliers:")
display(sa_gdp_out)

print("Middle East GDP Outliers:")
display(me_gdp_out)

```
